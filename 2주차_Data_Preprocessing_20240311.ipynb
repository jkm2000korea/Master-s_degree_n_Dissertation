```python
#5. Check and handle missing values
print(df.isnull())   #null 유무 확인
print(df.isnull().sum())   #변수별 null 개수 확인
print()

missing_value_rows = df[df.isnull().any(axis=1)]   #각 열을 따라서 하나라도 결측치가 있는 행을 출력
print(missing_value_rows)   #필터링된 행 출력
print()

missing_value_indexes = df[df.isnull().any(axis=1)].index   #결측치가 있는 행의 인덱스를 저장
df1 = df.fillna(1, inplace=False)   #결측치를 1로 채우고, 채워진 DataFrame을 df1로 저장 (inplace=False): 원데이터셋(df)는 건드리지 않음
filled_rows = df1.loc[missing_value_indexes]   #원래 결측치가 있었던 행만 불러오기
print(filled_rows)   #결과 출력
print()

df2 = df.dropna(inplace=False)   #결측치가 있는 행을 제거
remaining_indexes = df2.index   #df2에서 제거되지 않은 (즉, 남아 있는) 행의 인덱스를 가져옴
removed_indexes = missing_value_indexes.difference(remaining_indexes)   #원본 DataFrame의 결측치가 있었던 인덱스 중에서 제거된 인덱스만 추출
print(removed_indexes)   #결과 출력 (제거된 인덱스만 보여줌)
```

```python
#6. Check duplicated rows and values
print(df.duplicated('Region'))   #Region 변수 내 중복인 값을 True로 반환
print()
print(df.duplicated('Region').sum())   #Region 변수 내 중복 값을 가지는 관찰값의 수
print()
print(df.value_counts('Region'))   #각 중복데이터의 개수
print()
df3 = df.drop_duplicates('Region')   #중복데이터가 포함된 행 제거
print(df3)
```

```python
#7. Change values of certain variables into other type
#날짜 형식 변환
import datetime as dt

current_year = dt.datetime.now().year   #현시점의 연도를 저장

df['Birthdate'] = df['Birthdate'].astype(str).apply(
    lambda x: dt.datetime.strptime(f"{x.split('.')[0]}.{x.split('.')[1]}.{current_year - int(x.split('.')[2])}", "%m.%d.%Y")
    if len(x.split('.')) == 3 else pd.NaT   #3개의 요소, 즉 월.일.나이 형식을 따르지 않는 경우 결측된 값을 반환 (오류 예방)
)
df
```

```python
#7. Change values of certain variables into other type
#Age 내 음수값 양수값으로 변환
print(df.loc[df['Age'] < 0])   #Age 변수 내 음수값을 가지는 관찰값 확인
df['Age'] = df['Age'].abs()  #Age 변수 내 음수값을 양수값으로 변환
print(df.loc[df['Age'] < 0])   #Age 변수 내 음수값이 제거된 것 확인
```

```python
#8. Find problems and correct them
for i, column in df.iterrows():
  if df.at[i, 'Income'] < 9999:   #모든 행을 확인하면서 Income 변수의 값이 9999보다 작은 경우 아래를 수행
    df.at[i, 'Income'] = df.at[i, 'Income']*1000   #9999보다 작은 값에 1000을 곱하여 반환
df
```

```python
#9. Change values into suitable types for analysis
print(df.describe()[['Purchases']])   #Purchases 변수의 분포 확인
df_drop = df.drop(df.loc[df['Purchases']==9999].index, axis=0, inplace=False)    #Purchases 변수 내 9999값을 가지는 행 제거
print(df_drop.describe()[['Purchases']])    #제거 후 Purchases 변수의 분포 확인

df.drop(df_drop.loc[df_drop['Purchases' or 'AmountSpent']==0].index, axis=0, inplace=True)   #타겟 고객이 아닌 대상 제거(구매 또는 소비액=0)
df_add = df_drop.assign(Average = df_drop['AmountSpent']/df_drop['Purchases'])   #고객별 구매 시 평균 지출액 확인
df_add
```
